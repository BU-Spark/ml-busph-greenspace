## Green-space results Visualization

### SegForest

These are results generated by SegForest model, which are currently have the best results.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../results/school_50m_seg.csv')


# Sort the DataFrame by 'Area_50m_Tree_sqm' in descending order
df_sorted = df.sort_values(by='Area_50m_Tree_sqm', ascending=False)

# Select the top 10 rows
df_top10 = df_sorted.head(10)

# Create the bar plot
plt.figure(figsize=(10, 5))
plt.bar(df_top10['School_Name'], df_top10['Area_50m_Concrete_percent'], label='Concrete')
plt.bar(df_top10['School_Name'], df_top10['Area_50m_Building_percent'], bottom=df_top10['Area_50m_Concrete_percent'], label='Building')
plt.bar(df_top10['School_Name'], df_top10['Area_50m_LowVege_percent'], bottom=df_top10['Area_50m_Building_percent']+df_top10['Area_50m_Concrete_percent'], label='Low Vegetation')
plt.bar(df_top10['School_Name'], df_top10['Area_50m_Tree_percent'], bottom=df_top10['Area_50m_LowVege_percent']+df_top10['Area_50m_Building_percent']+df_top10['Area_50m_Concrete_percent'], label='Tree')
plt.bar(df_top10['School_Name'], df_top10['Area_50m_Water_percent'], bottom=df_top10['Area_50m_Tree_percent']+df_top10['Area_50m_LowVege_percent']+df_top10['Area_50m_Building_percent']+df_top10['Area_50m_Concrete_percent'], label='Water')
plt.xticks(rotation=90)
plt.xlabel('School Name')
plt.ylabel('Percentage')
plt.title('Percentage of Different Areas within 50 Meters of Top 10 Schools with Largest Area of Trees')
plt.legend()
plt.show()


plt.figure(figsize=(10, 5))
plt.scatter(df['Area_50m_LowVege_sqm'], df['Area_50m_Tree_sqm'])
plt.xlabel('Area of Low Vegetation (sqm)')
plt.ylabel('Area of Trees (sqm)')
plt.title('Area of Low Vegetation versus Area of Trees within 50 Meters of Each School')
plt.show()

plt.figure(figsize=(10, 5))
plt.boxplot([df['Area_50m_Concrete_sqm'], df['Area_50m_Building_sqm'], df['Area_50m_LowVege_sqm'], df['Area_50m_Tree_sqm'], df['Area_50m_Water_sqm']])
plt.xticks([1, 2, 3, 4, 5], ['Concrete', 'Building', 'Low Vegetation', 'Tree', 'Water'])
plt.ylabel('Area (sqm)')
plt.title('Area of Different Areas within 50 Meters of All Schools')
plt.show()

```
